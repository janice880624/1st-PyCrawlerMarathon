```
import requests
import pytesseract
from PIL import Image
from io import BytesIO
```

```
response = requests.get('http://pycrawler.cupoy.com/file-download/part32/test1.png')
img = Image.open(BytesIO(response.content))
img
```

```
code = pytesseract.image_to_string(img)
print(code)
```
